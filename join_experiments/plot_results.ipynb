```python
import json
gt_time = 1393.9
single_table_speedup = []
multi_table_speedup = []
with open("all_results.jsonl") as f:
    for line in f:
        data = json.loads(line)
        pilot_time = data["pilot_time"]
        single_table_time = data["single_table_time"]
        multi_table_time = data["multi_table_time"]
        single_table_speedup.append(gt_time / (single_table_time + pilot_time))
        multi_table_speedup.append(gt_time / (multi_table_time + pilot_time))
```

```python
import matplotlib.pyplot as plt
import numpy as np
# plot box plot
fig = plt.figure(figsize =(3, 5))
plt.boxplot([single_table_speedup, multi_table_speedup])
plt.xticks([1, 2], ['Single Table', 'Multi Table'])
plt.ylabel('Speedup over exact query time')
plt.savefig("speedup_boxplot.pdf", bbox_inches='tight')
```
